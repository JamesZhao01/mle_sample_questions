```python
x, y = torch.meshgrid(torch.arange(-12, 12, 0.1), torch.arange(-12, 12, 0.1))
k = len(x)
x, y = x.flatten(), y.flatten()

z = torch.sigmoid((x - y)/8) # change to decision function

x, y, z = x.view((k,k)), y.view((k,k)), z.view((k,k))
plt.pcolor(x.numpy(), y.numpy(), z.numpy(), cmap="plasma")
plt.scatter([1, 2, 3], [3, 2, 1], c=[0, 1, 2], s = [5] * 3, cmap="viridis")
plt.colorbar()
```
